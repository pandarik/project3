```python
import os

def init_logging_directory():

    log("init_logging_directory >>")
    
    sub_directory_path = "scape_data"

    # Check if the directory exists
    if not os.path.exists(sub_directory_path):
        # Create the directory if it does not exist
        os.makedirs(sub_directory_path)
        print(f"Directory '{sub_directory_path}' created.")
    else:
        print(f"Directory '{sub_directory_path}' already exists.")

    log("init_logging_directory <<")
```

```python
def log_soup_contents(url, soup_object, file_name):

    log("log_soup_contents >>")
    
    sub_directory_path = "scape_data"
    file_path = sub_directory_path + "/" + dt.datetime.now().strftime('%Y%m%d_%H%M%S') + "_" + file_name

    print(file_path)
    
    # Convert the Beautiful Soup object to a string
    soup_str = soup_object.prettify()

    # Write the string representation of the HTML content to the text file
    with open(file_path, 'w', encoding='utf-8') as file:
        file.write(url + '\n')
        file.write('----------------------------------------------------------------------------------------------------' + '\n')    
        file.write(soup_str + '\n')   

    log("log_soup_contents <<")
```

```python
import pandas as pd

def dataFrameComplete(rows):

    log("dataFrameComplete >>")

    # Convert the list of dictionaries to a DataFrame
    df = pd.DataFrame(rows)
    
    # Display the DataFrame
    print(df)
    
    log('First Row of Dataframe')
    log(df.iloc[0])

    # Write the DataFrame to a CSV file
    log('Writing Dataframe to CSV')
    sub_directory_path = "scape_data"
    file_name = 'dataframe_output.csv'
    file_path = sub_directory_path + "/" + dt.datetime.now().strftime('%Y%m%d_%H%M%S') + "_" + file_name
    df.to_csv(file_path, index=False)
    
    log("dataFrameComplete <<")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

log("main >>")

init_logging_directory()

# Initialize WebDriver (replace with the path to your WebDriver executable)
# driver = webdriver.Chrome('/path/to/chromedriver')
driver = webdriver.Chrome()

# Navigate to the main page
# main_page_url = 'https://example.com/main-page'
#main_page_url = 'http://www.google.com'
main_page_url = 'https://www.airbnb.com/s/Florida--United-States/homes'

log(f"main -- Fetching URL {main_page_url}")

driver.get(main_page_url)

log("main -- Driver Wait")

# Wait for the main page to load and be ready
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, 'body')))

# Optionally wait for specific content to be loaded if needed
log("main -- Sleeping 5 seconds")
time.sleep(5)  # Adjust this delay as needed

log("main -- Parsing Page Data")

# Get the main page source and parse it with BeautifulSoup
main_page_html = driver.page_source
main_soup = BeautifulSoup(main_page_html, 'html.parser')

log_soup_contents(main_page_url, main_soup, "soup_main_output.txt")



# #############################
# disabled this function , not working yet
# findSoupDiv18(main_soup)
# #############################



detail_url = parseMain(main_soup)

log(f"main -- Fetching URL {detail_url}")



# #############################
# time.sleep(60)  # Adjust this delay as needed
# #############################



# Navigate to the detail page
driver.get(detail_url)

log("main -- Driver Wait")

# Wait for the detail page to load and be ready
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, 'body')))

# Optionally wait for specific content to be loaded if needed
log("main -- Sleeping 5 seconds")
time.sleep(5)  # Adjust this delay as needed

log("main -- Parsing Page Data")

# Get the detail page source and parse it with BeautifulSoup
detail_page_html = driver.page_source
detail_soup = BeautifulSoup(detail_page_html, 'html.parser')

log_soup_contents(detail_url, detail_soup, "soup_detail_output.txt")

parseDetail(detail_soup)

# Now you can use BeautifulSoup to extract data from the detail page
# log(detail_soup.prettify())

# Close the browser
driver.quit()

log("main <<")
```
